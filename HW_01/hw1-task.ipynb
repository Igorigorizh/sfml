```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

#TODO -> Done
```

```python
# Функция преобразования значений LIMIT_BAL в буквенные значения
def LIMIT_BAL_RangeConvert2Letter(limit_bal_value):
    if limit_bal_value <=10000:
        return 'A'
    elif limit_bal_value > 10000 and limit_bal_value <=100000:
        return 'B'
    elif limit_bal_value > 100000 and limit_bal_value <=200000:
        return 'C'
    elif limit_bal_value > 200000 and limit_bal_value <=400000:
        return 'D'
    elif limit_bal_value > 400000 and limit_bal_value <=700000:
        return 'E'
    elif limit_bal_value > 700000:
        return 'F'

#[LIMIT_BAL_RangeConvert2Letter(a) for a in df['LIMIT_BAL']][:10]
# отображаем преобразованные значения в новый столбец 'LIMIT_BAL_2_letter'
df['LIMIT_BAL_2_letter'] = df['LIMIT_BAL'].map(LIMIT_BAL_RangeConvert2Letter)
```
